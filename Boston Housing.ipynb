```python
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader
```

```python
import pandas as pd
```

```python
train_ds_pd = pd.read_csv('train.csv', dtype = 'float32')
print(train_ds_pd)
```

```python
train_ds1= torch.Tensor(train_ds_pd.values)
print(train_ds1)
```

```python
inputs = train_ds1[:,0:14]
print(inputs.shape)
```

```python
import torch.nn as nn
```

```python
model = nn.Linear(14, 1)
print(model.weight)
print(model.bias)
```

```python
preds = model(inputs)
print(preds)
```

```python
targets = train_ds1[:,14:15]
print(targets.shape)
print(targets)
```

```python
train_ds = TensorDataset(inputs, targets)
```

```python
batch_size = 10
train_dl = DataLoader(train_ds, batch_size, shuffle = True)
```

```python
import torch.nn.functional as F
loss_fn = F.mse_loss
```

```python
loss = loss_fn(preds, targets)
print(loss)
```

```python
opt = torch.optim.SGD(model.parameters(), lr = 0.00000141)
```

```python
for xb, yb in train_dl:
    print(xb)
    print(yb)
    break
```

```python
def fit(num_epochs, model, loss_fn, opt):
    for i in range(num_epochs):
        for xb, yb in train_dl:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            opt.step()
            opt.zero_grad()
            
        if (i+1)%10 == 0:
            print('Epoch : [{}/{}], Loss = {:.4f}'.format(i+1, num_epochs, loss.item()))
                
            
```

```python
fit(100, model, loss_fn, opt)
```
